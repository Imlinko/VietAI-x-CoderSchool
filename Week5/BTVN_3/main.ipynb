# Task 1: Liệt kê các sản phẩm có doanh thu lớn hơn 100.000 Euro

```python
import pandas as pd
path = ".\dataset_3.csv"
df = pd.read_csv(path,encoding ="utf-8")
print(df)
```

```python
print(df.columns)
```

```python
df["revenue"] = df['units_sold']*df["price"]

print(df[df["revenue"] > 100000].reset_index())
```

# Task 2: Liệt kê các sản phẩm có units_sold nằm trong top 100, đồng thời nằm trong nhóm 10% sản phẩm có rating cao nhất

```python
df_top = df[df["units_sold"].isin(df.nlargest(n=100,columns="units_sold")["units_sold"]) &
   df["rating"].isin(df.nlargest(n = 153,columns="rating")["rating"])]
print(df_top)
```

# Task 3: Có phải sản phẩm càng được rating nhiều thì càng bán được nhiều sản phẩm ko

```python
print(df.sort_values(by=["rating","units_sold"],ascending=True))
```

# Task 4: Thống kê Top 50 sản phẩm nằm trong rating từ 3.0 → 4.0 bán chạy nhất

```python
df[(df["rating"] >= 3.0) & (df['rating'] <=4.0) ].nlargest(n=50,columns = "units_sold")
```

# Task 5: Có bao nhiêu sản phẩm có giá bán lớn hơn 50 EURO

```python
print(df[df["price"] > 50].value_counts().sum())
```
